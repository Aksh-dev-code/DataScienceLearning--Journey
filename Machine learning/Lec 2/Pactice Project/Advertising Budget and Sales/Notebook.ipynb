### 🔎 Insights from Regression Models

- **TV ads** are the strongest single predictor of Sales. Using only TV budget, the model explains about **67% of sales variation**.  
- **Radio ads** have some effect but much weaker (**26% of variation explained**).  
- **Newspaper ads** barely help at all (**only ~3% explained**).  
- When we use **all three features together (TV + Radio + Newspaper)**, the model becomes much stronger, explaining about **90% of the variation in Sales** with lower error.  
- This means **multiple regression is the best choice** for prediction.  
- In simple words:  
  - TV ads drive sales the most,  
  - Radio ads add extra help,  
  - Newspaper ads don’t matter much.
